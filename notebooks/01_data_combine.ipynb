```python
import pandas as pd
import numpy as np
from datetime import datetime
```

### Sources
1. Non-Sarcastic
    - Huffington Post
    - NPR
    - Guardian
    - Not The Onion
    - Atlantic
2. Sarcastic
    - Onion
    - National Report
    - Reductress

```python
huffington_abc = pd.read_csv('../data/HuffPost+abc.csv')
huffington_abc = huffington_abc[['source', 'headline', 'date', 'is_sarcastic']]
huffington_abc.columns = ['source', 'title', 'date', 'is_sarcastic']
huffington_abc.sample(n = 5)
```

```python
npr = pd.read_csv('../data/NPR 2016 to 2020.csv')
npr['is_sarcastic'] = 0

npr['date_dt'] = pd.to_datetime(npr['date'], format='%B-%d-%Y')
npr['date'] = npr['date_dt'].apply(lambda x:str(x)[:10])

npr = npr[['source', 'title', 'date', 'is_sarcastic']]
npr.sample(n = 5)
```

```python
guardian = pd.read_csv('../data/Guardian 2017 to 2020.csv')
guardian['is_sarcastic'] = 0
guardian['date'] = pd.to_datetime(guardian['date'])
guardian['date'] = guardian['date'].apply(lambda x:str(x)[:10])


guardian = guardian[['source', 'title','date','is_sarcastic']]


guardian.sample(n = 5)
```

```python
nottheonion = pd.read_csv('../data/nottheonion_2020.csv')

nottheonion['is_sarcastic'] = 0
nottheonion['date'] = nottheonion['date'].apply(lambda x: datetime.utcfromtimestamp(int(x)).date())
nottheonion['date'] = nottheonion['date'].apply(lambda x:str(x)[:10])

nottheonion = nottheonion[['source', 'title', 'date', 'is_sarcastic']]

nottheonion.sample(n = 5)
```

```python
national_report = pd.read_csv('../data/national_report.csv')
national_report['source'] = 'national report'
national_report['is_sarcastic'] = 1

national_report['date'] = pd.to_datetime(national_report.Dates)
national_report['date'] = national_report['date'].apply(lambda x:str(x)[:10])

national_report = national_report[['source', 'Headlines', 'date', 'is_sarcastic']]
national_report.columns = ['source', 'title', 'date', 'is_sarcastic']

national_report.sample(n = 5)
```

```python
atlantic = pd.read_csv('../data/atlantic_cleaned.csv')
atlantic['source'] = 'atlantic'
atlantic['is_sarcastic'] = 0

atlantic = atlantic[['source', 'headline', 'date', 'is_sarcastic']]
atlantic.columns = ['source', 'title', 'date', 'is_sarcastic']

atlantic.sample(5)
```

```python
reductress = pd.read_csv('../data/reductress_cleaned.csv')
reductress['is_sarcastic'] = 1
reductress['source'] = 'reductress'
reductress['date'] = reductress['year'].apply(lambda x: str(x) + '-01-01')

reductress = reductress[['source','headline','date','is_sarcastic']]
reductress.columns = ['source','title','date','is_sarcastic']

reductress.sample(n = 5)
```

```python
onion = pd.read_csv('../data/onion_headlines.csv')
onion['date'] = pd.to_datetime(onion['date'])
onion['date'] = onion['date'].apply(lambda x:str(x)[:10])
onion['is_sarcastic'] = 1

onion = onion[['source','headlines','date','is_sarcastic']]
onion.columns =  ['source','title','date','is_sarcastic']

onion.sample(n = 5)
```

```python
all_dfs = [huffington_abc, npr, guardian, nottheonion, atlantic, onion, reductress, national_report]
combined_df = pd.concat(all_dfs, axis = 0)

combined_df.columns = ['source', 'headline', 'date', 'is_sarcastic']
combined_df['year'] = combined_df['date'].apply(lambda x: x[:4])

combined_df = combined_df[['source', 'headline', 'year', 'date', 'is_sarcastic']]

combined_df.sample(n = 10)
```

```python
combined_df.to_csv('../data/combined_data_v3.csv', index = False)
```
