```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
res = requests.get('https://www.theguardian.com/us-news?page=1900')
soup_data = BeautifulSoup(res.text, 'html.parser')
#soup_data.find_all(class_='u-faux-block-link__overlay js-headline-text')
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; rv:20.0) Gecko/20100101 Firefox/20.0'}

guardian_title = []
guardian_timestamp = []

for page in range(1,1901):
    res = requests.get(f'https://www.theguardian.com/us-news?page={page}', headers = headers)
    each_page = BeautifulSoup(res.text, 'html.parser')
    title = each_page.find_all(class_='u-faux-block-link__overlay js-headline-text')
    timestamp = each_page.find_all(class_ = "fc-item__timestamp") 
    
    all_titles=[]
    for e in title:
        all_titles.append(e.get_text())
    
    all_timestamps=[]
    for t in timestamp:
        all_timestamps.append(int(t['data-timestamp']))
        all_timestamps = all_timestamps[:len(all_titles)]
    
    guardian_title.append(all_titles)
    guardian_timestamp.append(all_timestamps)
```

```python
t_df = pd.DataFrame(np.transpose(guardian_title))
title_df = pd.concat([t_df, t_df.T.stack().reset_index(name='title')['title']], axis=1)
title_df = title_df['title']
title_df = pd.DataFrame(title_df)
title_df
```

```python
t_df = pd.DataFrame(np.transpose(guardian_timestamp))
time_df = pd.concat([t_df, t_df.T.stack().reset_index(name='timestamp')['timestamp']], axis=1)
time_df = time_df['timestamp'].astype(str)
time_df = pd.DataFrame(time_df)
time_df
```
