### Final Data 

```python
nums = [i for i in range(1,1295)]
atlantic_headlines =  list()
atlantic_dates = list()
for num in nums:
    page = requests.get('https://www.theatlantic.com/latest/?page={}'.format(num))
    soup = BeautifulSoup(page.text, 'html.parser')
    #Headline Reader
    headlines = soup.find_all(class_='hed')
    for i in range(len(headlines)):
        atlantic_headlines.append(str(headlines[i]).split('\n')[1].strip())
    #Date Reader 
    dates = soup.find_all(class_='date')
    for i in range(len(dates)):
        x = str(dates[i])
        y = re.search('[2][0][1-2][0-9]-[0-9][0-9]-[0-9][0-9]',x).group()
        atlantic_dates.append(y)
atlantic_headlines
```

```python
frame = {'Headlines':pd.Series(atlantic_headlines), 'Dates':pd.Series(atlantic_dates)}
atlantic_df = pd.DataFrame(frame)
atlantic_df.to_csv('atlantic_csv',index=False)
```

## Reductress  

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; rv:20.0) Gecko/20100101 Firefox/20.0'}
page = requests.get(f'https://reductress.com/page/100/?s', headers = headers)
soup = BeautifulSoup(page.text, 'html.parser')
```

### Headline Reader

```python
headlines = soup.find_all('h1')
headlines_filtered = list()
extraneous_lines = ['<h1>SEARCH RESULTS FOR ‘’</h1>', '<h1>POPULAR:</h1>',
                    '<h1>REDUCTRESS ON <i class="fab fa-twitter"></i></h1>','<h1>Read More:</h1>',
                    '<h1>Follow Us:</h1>', '<h1>Sign Up For Our Newsletter...</h1>']
for line in headlines:
    if str(line) not in extraneous_lines:
        temp = str(line)[4:][0:-5]
        headlines_filtered.append(temp)
headlines_filtered
```

### Reductress Final Loop 


```python
reductress_headlines = list()
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; rv:20.0) Gecko/20100101 Firefox/20.0'}
for i in range(1,388):
    page = requests.get('https://reductress.com/page/{}/?s'.format(i), headers = headers)
    soup = BeautifulSoup(page.text, 'html.parser')
    headlines = soup.find_all('h1')
    extraneous_lines = ['<h1>SEARCH RESULTS FOR ‘’</h1>', '<h1>POPULAR:</h1>',
                    '<h1>REDUCTRESS ON <i class="fab fa-twitter"></i></h1>','<h1>Read More:</h1>',
                    '<h1>Follow Us:</h1>', '<h1>Sign Up For Our Newsletter...</h1>']
    for line in headlines:
        if str(line) not in extraneous_lines:
            temp = str(line)[4:][0:-5]
            reductress_headlines.append(temp)
```

```python
frame = {'Headlines':pd.Series(reductress_headlines)}
reductress_df = pd.DataFrame(frame)
reductress_df.to_csv('reductress_csv',index=False)
```

## The National Report (Satirical)

### Headline Reader

```python
#page = requests.get(f'https://nationalreport.net/page/1/?s=+&x=0&y=0')
soup = BeautifulSoup(page.text, 'html.parser')
soup
```

```python
headlines = soup.find_all('h2')
headlines_filtered = list()
for line in headlines:
    x = str(line)
    y = re.search('[P][e][r][m][a][l][i][n][k].*"',x).group()
    temp = y[13:][0:-1].strip()
    headlines_filtered.append(temp)
headlines_filtered
```

### Date Reader 

```python
dates = soup.find_all(class_="entry-meta")
dates_filtered = list()
for date in dates:
    x = str(date)
    y = re.search('[JFMASOND][a-z]*\s[0-3][0-9].*[2][0][1-2][0-9]',x).group()
    dates_filtered.append(y)
dates_filtered
```

### Final Data 

```python
pages = [i for i in range(1,8)]
national_report_headlines = list()
for page in pages:
    webpage = requests.get('https://nationalreport.net/page/{}/?s=+&x=0&y=0'.format(page))
    soup = BeautifulSoup(webpage.text, 'html.parser')
    #Headline Reader 
    headlines = soup.find_all('h2')
    for line in headlines:
        x = str(line)
        y = re.search('[P][e][r][m][a][l][i][n][k].*"',x).group()
        temp = y[13:][0:-1].strip()
        national_report_headlines.append(temp)  
national_report_headlines
```

```python
national_report_dates = list()
for page in pages:
    webpage = requests.get('https://nationalreport.net/page/{}/?s=+&x=0&y=0'.format(page))
    soup = BeautifulSoup(webpage.text, 'html.parser')
    #Date Reader
    dates = soup.find_all(class_="entry-meta")
    for date in dates:
        x = str(date)
        y = re.search('[JFMASOND][a-z].*[2][0][1-2][0-9]',x).group()
        national_report_dates.append(y)
national_report_dates
```

```python
national_report_headlines_series = pd.Series(national_report_headlines)
national_report_dates_series = pd.Series(national_report_dates)
frame = {'Headlines':national_report_headlines_series, 'Dates':national_report_dates_series}
national_report_df = pd.DataFrame(frame)
national_report_df.to_csv('national_report_csv',index=False)
```
