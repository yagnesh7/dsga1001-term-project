```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime
import time
import numpy as np
```

```python
music_years = {
    '2016': [124,152],
    '2017': [95,124],
    '2018': [69,94],
    '2019': [27,68],
    '2020': [1,26]
}

# https://thehardtimes.net/music/page/169/

gaming_years = {
    '2016': [136,136],
    '2017': [119,135],
    '2018': [93,119],
    '2019': [50,92],
    '2020': [1,49]
}

# https://thehardtimes.net/harddrive/page/49/
```

```python
entries = []

for year in music_years.keys():
    print(year)
    start = music_years[year][0]
    end = music_years[year][1]
    for i in range(start, end+1):
        print(i)
        if i == 0:
            continue
        headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; rv:20.0) Gecko/20100101 Firefox/20.0'}
        page = requests.get(f'https://thehardtimes.net/music/page/{i}', headers = headers)
        soup = BeautifulSoup(page.text, 'html.parser')
        headlines = soup.find_all(class_="post-title")
        headline_text = [x.text.replace('\n','') for x in headlines]
        for entry in headline_text:
            result = {'source': 'Hard Times', 'year': str(year), 'date': f'{year}-01-01', 'title': entry}
            entries.append(result)
            
        time.sleep(np.random.uniform(.5, 1.5))
```

```python
hard_time_music = pd.DataFrame(entries)
print(hard_time_music.shape)
hard_time_music.sample(5)
```

```python
entries = []

for year in gaming_years.keys():
    print(year)
    start = gaming_years[year][0]
    end = gaming_years[year][1]
    for i in range(start, end+1):
        print(i)
        if i == 0:
            continue
        headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; rv:20.0) Gecko/20100101 Firefox/20.0'}
        page = requests.get(f'https://thehardtimes.net/harddrive/page/{i}', headers = headers)
        soup = BeautifulSoup(page.text, 'html.parser')
        headlines = soup.find_all(class_="post-title")
        headline_text = [x.text.replace('\n','') for x in headlines]
        for entry in headline_text:
            result = {'source': 'Hard Times', 'year': str(year), 'date': f'{year}-01-01', 'title': entry}
            entries.append(result)
            
        time.sleep(np.random.uniform(.5, 1.5))
```

```python
hard_time_gaming = pd.DataFrame(entries)
print(hard_time_gaming.shape)
hard_time_gaming.sample(5)
```
