```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime
import time
import numpy as np
import json
import re
import time
```

```python
def response_parse(items_list):
    return [{'source': item['pagemap']['metatags'][0]['dc.source'],
            'date': item['pagemap']['metatags'][0]['dc.date'],
             'title': item['pagemap']['metatags'][0]['og:title']} for item in items_list]
```

```python
words = pd.read_table('common_words.txt', header=None)[0]
```

```python
results = []

for word in words[446:]:
    print(word)
    
    for year in range(2016,2021):
        
        url = f"https://api.foxnews.com/search/web?q={word}+-filetype:amp+-filetype:xml+more:pagemap:metatags-prism.section&siteSearch=foxnews.com&siteSearchFilter=i&sort=date:r:{year}0101:{year}1231&callback=__jp6"
        r = requests.get(url)
        
        if r.status_code != 200:
            print("Bad Call")
            print(r.status_code)
            continue
        
        
        try:
            s = str(r.content)
            s = s[s.index("(") + 1:-(len(s) - s.index(");"))]
            s = s.replace("\\n","").replace('\\\\"', "'").replace("\\", "")
            d = json.loads(s)
            
            if int(d['searchInformation']['totalResults']) > 0:
                results.extend(response_parse(d['items']))
            else:
                print("Empty Results")
            
        except:
            print(f"Failed: {word} for {year}")
    
    print(len(results))
    pd.DataFrame(results).drop_duplicates().to_csv('fox_news_4.csv', index = False)
    time.sleep(np.random.uniform(low = 0.25, high = 0.75))
```
